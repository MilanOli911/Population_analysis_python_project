Function for calculating the Population Growth

```python
def pop_growth_stats(year1,year2):
  pop_year1 = population_data1.loc[population_data1['Year']== year1]
  pop_year2 = population_data1.loc[population_data1['Year']== year2]
  pop_growth3 = pop_year1.merge(pop_year2 , left_on = 'Country Name',right_on = 'Country Name')
  pop_growth3 = pop_growth3.drop(['Year_x','Year_y'],axis = 1)
  pop_growth3.rename(columns = {'population_x':'population_year1','population_y':'population_year2'},inplace = True)
  pop_growth3['per_growth1'] =100 * ( pop_growth3['population_year2'] - pop_growth3['population_year1'])/pop_growth3['population_year1']
  print("Country having Maximum Growth rate")
  print(pop_growth3.nlargest(n=1,columns=['per_growth1']))
  print("Country having Minimum Growth rate")
  print(pop_growth3.nsmallest(n=1,columns=['per_growth1']))
  

```

Method Calling

```python
pop_growth_stats(year1,year2)
```
